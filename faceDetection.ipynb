```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image
img=cv2.imread("kidsImg.png")
print(img)
```

```python
plt.imshow(img)
```

```python
#convert color
newimg=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(newimg)
plt.axis("off")
```

```python
model=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=model.detectMultiScale(newimg,1.2,3)
```

```python
faces
```

```python
for f in faces:
    x,y,w,h=f
    cv2.rectangle(newimg,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(newimg)
plt.show()    
```
